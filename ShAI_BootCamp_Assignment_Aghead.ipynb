Data Cleaning:

```python
df['Agency'].nunique()
```

```python
df = df.drop(columns = ['Notes' , 'Status' , 'Agency'])
#deleting Agency until new agency added
```

```python
df.tail(33)
```

```python
df = df.iloc[:-33]
print (df.isnull().sum())
```

```python
filtered_benefits = df[df['Benefits'] > 0]
most_frequent_value = filtered_benefits['Benefits'].mode()[0]
print (most_frequent_value)

df['BasePay'] = df['BasePay'].fillna(method='ffill')
df['Benefits'] = df['Benefits'].fillna(most_frequent_value)
```

```python
print (df.isnull().sum())
```

Basic Data Visualization

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(df['BasePay'], bins=30, edgecolor='black')
plt.title('Histogram of BasePay')
plt.xlabel('BasePay')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['Benefits'], bins=30, edgecolor='black')
plt.title('Histogram of Benefits')
plt.xlabel('Benefits')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['TotalPay'], bins=30, edgecolor='black')
plt.title('Histogram of TotalPay')
plt.xlabel('TotalPay')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
jobtitle_frequencies = df['JobTitle'].value_counts()

# Select the top 10 most frequent job titles
top_jobtitles = jobtitle_frequencies[:10]

# Create the pie plot
plt.figure(figsize=(10, 10))
top_jobtitles.plot.pie(autopct='%1.1f%%', figsize=(10, 10))
plt.title('Top 10 Most Frequent Job Titles')
plt.show()
```

Grouped Analysis:

```python
interest_columns= ['JobTitle', 'Year']

# Perform grouped analysis for each column using a for loop
for column in interest_columns:
    grouped_by_column = df.groupby(column)
    average_salary_by_column = grouped_by_column['TotalPay'].mean(0)
    print(f"Average Salary by {column}:\n", average_salary_by_column)
    print("\n")
```

Simple Correlation Analysis

```python

# Calculate the correlation between 'TotalPay' and another numerical column
correlation = df['TotalPay'].corr(df['OvertimePay'])

# Print the correlation coefficient
print(f"Correlation between TotalPay and OvertimePay: {correlation}")

# Create a scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
plt.scatter(df['TotalPay'], df['OvertimePay'], alpha=0.5)
plt.title('TotalPay vs. OvertimePay')
plt.xlabel('TotalPay')
plt.ylabel('OvertimePay')
plt.show()
```
